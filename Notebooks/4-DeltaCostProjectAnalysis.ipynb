We define a function to search through the list of colleges in the database and produce a list of indices corresponding to open admissions 4-year public and private non-profit institutions with at least 100 students.

```python
def search_colleges(dataframe):
    df = dataframe
    df.fillna(-1)
    college_name_list = []
    college_index_list = []
    for x in range(len(df.index)):
        if df['admitcount'].iloc[x] >= 100:
            if (df['applicantcount'].iloc[x] / df['admitcount'].iloc[x]) == 1 and (df['sector_revised'].iloc[x] == 1 or df['sector_revised'].iloc[x] == 2):
                if df['instname'].iloc[x] not in college_name_list:
                    college_name_list.append(df['instname'].iloc[x])
                    college_index_list.append(x)
    return college_index_list
```

We can then use the indices returned in this function to create a new dataframe with only our desired institutions.

```python
open_admission_df = delta.iloc[search_colleges(delta)]
display(open_admission_df)
```
